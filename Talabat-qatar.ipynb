```python
import requests
import pandas as pd
```

```python
data=[]

for i in range(1,19):
  url='https://www.talabat.com/qatar/restaurants?page={}'.format(i)
  response= requests.get(url)
  soup= BeautifulSoup(response.text, 'lxml')
  restaurants= soup.find_all('div', class_='vendorstyles__VendorContainer-eu6e2y-0 hAgray mx-2 mb-4')
  for res in restaurants:
    resturant=[]
    children = res.findChildren("p")

    resturant.append(children[0].text) 
    resturant.append(children[1].text) 
    data.append(resturant)
data


```

```python

df = pd.DataFrame(data, columns =['restaurant Name', 'Category'])
df
```
